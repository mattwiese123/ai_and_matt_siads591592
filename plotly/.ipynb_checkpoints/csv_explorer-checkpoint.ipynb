```python
import os
import pandas as pd

csv_files = {}
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        if '.csv' in filename:
            print(os.path.join(dirname, filename))
            csv_files[filename] = pd.read_csv(os.path.join(dirname, filename))
```

```python
for file in csv_files:
    print('============================')
    print(file)
    print('============================\n')
    print(csv_files[file].head())
    print('####################\n####################\n####################\n')
```

```python
import pickle
import pandas as pd

data = pickle.load(open('vaccine_trends.pkl', 'rb'))
data2 = pickle.load(open('covid_trends.pkl', 'rb'))
```

```python
vaccine_ts = pd.DataFrame(data['trends_interest_over_time']['data'])
vaccine_ts['value'] = vaccine_ts['values'].apply(lambda x: x[0]['value'])
vaccine_ts['date_utc'] = pd.to_datetime(vaccine_ts['date_utc'])
```

```python
covid_ts = pd.DataFrame(data2['trends_interest_over_time']['data'])
covid_ts['has_values'] = covid_ts['values'].apply(lambda x: x[0]['has_value'])
covid_ts=covid_ts[covid_ts['has_values']==True]
covid_ts['value'] = covid_ts['values'].apply(lambda x: x[0]['value'])
covid_ts['date_utc'] = pd.to_datetime(covid_ts['date_utc'])
```

```python
%matplotlib inline
vaccine1 = vaccine_ts[['date_utc', 'value']].set_index('date_utc')
```

```python
covid1 = covid_ts[['date_utc', 'value']].set_index('date_utc')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,5))
plt.plot(vaccine1, label='vaccine')
plt.plot(covid1, label='covid')
plt.title('Search Trends on Google')
plt.legend();
```
